```python
with open('key.json') as f:
	key = json.load(f)
	bearer_token = key['BEARER_TOKEN']
```

```python
# Json path
json_path = 'followers.json'
with open(json_path, 'r') as f:
	follower_json = json.load(f)
```

```python
# Get timeline
def get_timeline(user_id, num_tweets = 5):
	# Create the url
	url = 'https://api.twitter.com/2/users/' + user_id + '/tweets?max_results=' + str(num_tweets)
	# Create request headers
	headers = {
		'authorization': 'Bearer ' + bearer_token,
		'user-agent': 'v2FilteredStreamPython',
	}
	# Make the request
	response = requests.get(url, headers=headers)
	return response.json()
```

```python
# Create list of follower ids
follower_time_line = {}
for user in follower_json:
	for follower in follower_json[user]["data"]:
		try:
			# Get followers timeline
			timeline = get_timeline(follower['id'], 100)
			# Make sure that follower_time_line is not empty
			if not follower_time_line[user]:
				follower_time_line[user] = {}
			follower_time_line[user][follower['id']] = timeline
			print(json.dumps(timeline, indent=4))
		except:
			# If the user has no timeline, add an empty list
			follower_time_line[user] = {}	
follower_alias_json["followers_timeline"] = follower_time_line
```

```python
# Write to json
with open('followers_timeline.json', 'w') as f:
	json.dump(follower_alias_json, f, indent=4)
```
