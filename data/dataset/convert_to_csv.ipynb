```python
from analysis import *
import random
```

```python
bad_tweet_list = analysis.create_tweet_list_from_loaded_json(analysis.universal_data_loader("posable_bad_tweets.json"))
good_tweet_list = analysis.create_tweet_list_from_loaded_json(analysis.universal_data_loader("all_good_tweets_v4.json"))
false_positive_list = analysis.create_tweet_list_from_loaded_json(analysis.universal_data_loader("false_postives_from_v3_ivory.json"))
```

```python
print("bad_tweet_list:", len(bad_tweet_list))
print("good_tweet_list:", len(good_tweet_list))
print("false_positive_list:", len(false_positive_list))
```

```python
for tweet in bad_tweet_list:
    # Get just the tweet's text
    tweet_text = analysis.get_tweet_text(tweet)

    # Remove all url's from tweet and replace with 'URL'
    tweet_text = analysis.replace_urls(tweet_text)

    # Replace all @username
    tweet_text = analysis.replace_mentions(tweet_text)
	
	# Excape any quotes in the tweet
    tweet_text = tweet_text.replace('"', '""')
    
    # Replace the tweet text
    tweet["text"] = tweet_text


# Remove all the tweets that are not in english
for tweet in good_tweet_list:
        # Get just the tweet's text
    tweet_text = analysis.get_tweet_text(tweet)

    # Remove all url's from tweet and replace with 'URL'
    tweet_text = analysis.replace_urls(tweet_text)

    # Replace all @username
    tweet_text = analysis.replace_mentions(tweet_text)
	

    
    # If the tweet is not in english, remove it
    if not analysis.is_english(tweet_text):
        print("Removing tweet:", tweet_text)
        good_tweet_list.remove(tweet)
    else:
        # Excape any quotes in the tweet
        tweet_text = tweet_text.replace('"', '""')
        
        # Replace the tweet text
        tweet["text"] = tweet_text
    
for tweet in false_positive_list:
        # Get just the tweet's text
    tweet_text = analysis.get_tweet_text(tweet)

    # Remove all url's from tweet and replace with 'URL'
    tweet_text = analysis.replace_urls(tweet_text)

    # Replace all @username
    tweet_text = analysis.replace_mentions(tweet_text)
	

    
    # If the tweet is not in english, remove it
    if not analysis.is_english(tweet_text):
        print("Removing tweet:", tweet_text)
        false_positive_list.remove(tweet)
    else:
        # Excape any quotes in the tweet
        tweet_text = tweet_text.replace('"', '""')
        
        # Replace the tweet text
        tweet["text"] = tweet_text

print("Bad tweets:", len(bad_tweet_list))
print("Good tweets:", len(good_tweet_list))
print("False positives:", len(false_positive_list))
```

```python
tweet_list = bad_tweet_list + good_tweet_list + false_positive_list
```

```python
print("Total tweets:", len(tweet_list))
print(analysis.get_tweet_text(tweet_list[1]))
```

```python
# Set the seed
random.seed(42)

# Shuffle the tweet list
random.shuffle(tweet_list)

# Write to CSV
tweet_texts = []
tweet_id = []
fake_id = []
for tmp_fake_id, tweet in enumerate(tweet_list):
    
    tweet_text = analysis.get_tweet_text(tweet)
    
    # Append text to list
    tweet_texts.append(tweet_text)
    
    # Append tweet id to list
    tweet_id.append(analysis.get_tweet_id(tweet))
    
    # Append fake id to list
    fake_id.append(tmp_fake_id)
    
with open("tweets.csv", "w") as f:
    for tweet_id, tweet_text in enumerate(tweet_texts):
        # Write to CSV
        f.write('"' + str(tweet_id) + '","' + tweet_text + '" \n')
        
with open("tweet_ids_to_fake_ids.csv", "w") as f:
    for fake_id, tweet in enumerate(tweet_list):
        try:
            f.write('"' + str(fake_id)+ '","' + str(analysis.get_tweet_id(tweet))  + '" \n')
        except:
            f.write('"' + str(fake_id) + '","' + "No tweet id" + '" \n') 
```
