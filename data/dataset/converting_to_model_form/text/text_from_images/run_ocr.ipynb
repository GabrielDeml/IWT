```python
import pytesseract
import cv2 
import glob
import json
```

```python
# Run OCR on all images in the folder
ids_to_text = {}
for filename in glob.glob('images_fake_ids_final/*'):
	img = cv2.imread(filename)
	text = pytesseract.image_to_string(img)

	tweet_id = filename.split('/')[-1].split('.')[0].split('_')[0]
	print(filename, tweet_id)
	if tweet_id in ids_to_text.keys():
		ids_to_text[tweet_id].append(text)
	else:
		ids_to_text[tweet_id] = [text]
	# ids_to_text[id] = text
	# print(text)

for tweet_id in ids_to_text.keys():
	tweet = ids_to_text[tweet_id]
	tweet = ' '.join(tweet)
	# Remove all url's from tweet and replace with 'URL'
	tweet = analysis.replace_urls(tweet)

    # Replace all @username
	tweet = analysis.replace_mentions(tweet)
    
    # Replace all emails
	tweet = analysis.replace_emails(tweet)
	ids_to_text[tweet_id] = tweet
	
```

```python
# Write json
with open('ids_to_text.json', 'w') as f:
	json.dump(ids_to_text, f)
```
