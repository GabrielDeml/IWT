```python
tweets_list_possable_bad_tmp = []
for tweet in tweets_list_possable_bad:
    # Get just the tweet's text
    tweet_text = analysis.get_tweet_text(tweet)

    # Remove all url's from tweet and replace with 'URL'
    tweet_text = analysis.replace_urls(tweet_text)

    # Replace all @username
    tweet_text = analysis.replace_mentions(tweet_text)
	
	# Excape any quotes in the tweet
    tweet_text = tweet_text.replace('"', '""')
    
    # Replace the tweet text
    tweet["text"] = tweet_text
    
    tweets_list_possable_bad_tmp.append(tweet)


tweets_list_good_tmp = []
# Remove all the tweets that are not in english
for tweet in tweets_list_good:
        # Get just the tweet's text
    tweet_text = analysis.get_tweet_text(tweet)

    # Remove all url's from tweet and replace with 'URL'
    tweet_text = analysis.replace_urls(tweet_text)

    # Replace all @username
    tweet_text = analysis.replace_mentions(tweet_text)
    
    # If the tweet is not in english, remove it
    try:
        if not analysis.is_english(tweet_text):
            print("Removing tweet:", tweet_text)
            # tweets_list_good.remove(tweet)
        else:
            # Excape any quotes in the tweet
            tweet_text = tweet_text.replace('"', '""')
            
            # Replace the tweet text
            tweet["text"] = tweet_text
            
            # Add the tweet to the list
            tweets_list_good_tmp.append(tweet)
    except:
        # If the tweet is not in english, remove it
        # tweets_list_good.remove(tweet)
        print("Removing tweet:", tweet_text)

tweets_list_false_positives_tmp = [] 
for tweet in tweets_list_false_positives:
        # Get just the tweet's text
    tweet_text = analysis.get_tweet_text(tweet)

    # Remove all url's from tweet and replace with 'URL'
    tweet_text = analysis.replace_urls(tweet_text)

    # Replace all @username
    tweet_text = analysis.replace_mentions(tweet_text)
    try:
        # If the tweet is not in english, remove it
        if not analysis.is_english(tweet_text):
            print("Removing tweet:", tweet_text)
            # tweets_list_false_positives.remove(tweet)
        else:
            # Excape any quotes in the tweet
            tweet_text = tweet_text.replace('"', '""')
            
            # Replace the tweet text
            tweet["text"] = tweet_text
            
            # Add the tweet to the list
            tweets_list_false_positives_tmp.append(tweet)
    except:
        # If the tweet is not in english, remove it
        # tweets_list_false_positives.remove(tweet)
        print("Removing tweet:", tweet_text)

seed_user_good_tweets_tmp = []
for tweet in seed_user_good_tweets:
        # Get just the tweet's text
    tweet_text = analysis.get_tweet_text(tweet)

    # Remove all url's from tweet and replace with 'URL'
    tweet_text = analysis.replace_urls(tweet_text)

    # Replace all @username
    tweet_text = analysis.replace_mentions(tweet_text)
    try:
        # If the tweet is not in english, remove it
        if not analysis.is_english(tweet_text):
            print("Removing tweet:", tweet_text)
            # seed_user_good_tweets.remove(tweet)
        else:
            # Excape any quotes in the tweet
            tweet_text = tweet_text.replace('"', '""')
            
            # Replace the tweet text
            tweet["text"] = tweet_text
            
            # Add the tweet to the list
            seed_user_good_tweets_tmp.append(tweet)
    except:
        # If the tweet is not in english, remove it
        # seed_user_good_tweets.remove(tweet)
        print("Removing tweet:", tweet_text)

print("bad_tweet_list:", len(tweets_list_possable_bad_tmp))
print("good_tweet_list:", len(tweets_list_good_tmp))
print("false_positive_list:", len(tweets_list_false_positives_tmp))
print("seed_user_good_tweets:", len(seed_user_good_tweets_tmp))
```

```python
# Get all image paths from the image folder
image_id_to_path = {}
for image in glob.glob("images/*"):
    # Get the image id
	image_id = image.split("/")[1]
	# Remove everything after _x.jpg
	image_id = image_id.split("_")[0]
 
	# Check if the image id is in the image_id_to_path dictionary
	if image_id in image_id_to_path:
		image_id_to_path[image_id].append(image)
	else:
		# Add the image id to the image path
		image_id_to_path[image_id] = [image]
	
```

```python
# Set the seed
random.seed(42)

# Shuffle the tweet list
random.shuffle(tweet_list)

# Write to CSV
tweet_texts = []
tweet_id = []
fake_id = []
for tmp_fake_id, tweet in enumerate(tweet_list):

    tweet_text = analysis.get_tweet_text(tweet)

    # Append text to list
    tweet_texts.append(tweet_text)

    # Append tweet id to list
    tweet_id.append(analysis.get_tweet_id(tweet))

    # Append fake id to list
    fake_id.append(tmp_fake_id)

    # Rename to image file to the fake id
    if analysis.get_tweet_id(tweet) in image_id_to_path:
        image_paths = image_id_to_path[analysis.get_tweet_id(tweet)]
        counter = 0
        for image_path in image_paths:
            # Get extension
            extention = image_path.split(".")[-1]
            shutil.copy(image_path, "images_fake_ids/" + str(tmp_fake_id) + "_" + str(counter) + "." + extention)
            counter += 1


with open("tweets.csv", "w") as f:
    for tweet_id, tweet_text in enumerate(tweet_texts):
        # Write to CSV
        f.write('"' + str(tweet_id) + '","' + tweet_text + '" \n')

with open("tweet_ids_to_fake_ids.csv", "w") as f:
    for fake_id, tweet in enumerate(tweet_list):
        try:
            f.write('"' + str(fake_id) + '","' +
                    str(analysis.get_tweet_id(tweet)) + '" \n')
        except:
            f.write('"' + str(fake_id) + '","' + "No tweet id" + '" \n')
```
