```python
import requests
import json
import time
import random
```

```python
# Get BEARER_TOKEN from key.json
with open('../../key.json') as f:
	key = json.load(f)
	bearer_token = key['BEARER_TOKEN']
```

```python
# Json path
json_path = '../../GMouYes/data/userProfile.json'
# Open json file
json_file = open(json_path, 'r')
# Load json file
json_data = json.load(json_file)
```

```python
# Get random user id
def random_id():
	# Get random number between 0 and 9
	user_id = ""
	user_id_length = random.randint(1,19)
	for i in range(user_id_length):
		user_id += str(random.randint(0,9))
	return user_id
```

```python
# Get timeline
def get_timeline(num_tweets = 5):
	# Create the url
	while True:
		user_id = random_id()
		url = 'https://api.twitter.com/2/users/' + user_id + '/tweets?max_results=' + str(num_tweets)
		# Create request headers
		headers = {
			'authorization': 'Bearer ' + bearer_token,
			'user-agent': 'v2FilteredStreamPython',
		}
		response = requests.get(url, headers=headers).json()
		try:
			if response["errors"][0]["value"] == user_id:
				print("Bad user id")

			else:
				break
		except:
			break
	
	return response, user_id
```

```python
users_timeline = {}
for i in range(100):
	# Get the timeline 
	timeline, user_id = get_timeline(100)
	# Add the timeline to the dictionary
	users_timeline[user_id] = timeline
# Write the dictionary to a json file
with open('../../data/random_users_timeline.json', 'w') as f:
	json.dump(users_timeline, f)
```
