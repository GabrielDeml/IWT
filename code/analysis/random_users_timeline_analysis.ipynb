```python
import json
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import os
import requests
```

```python
followers_analysis_json = {}
```

```python
# Json path
json_path = '../../data/random_users_timeline.json'
```

```python
# Open json file
followers_timeline_file = open(json_path, 'r')
# Load json file
followers_timeline = json.load(followers_timeline_file)
```

```python
print(json.dumps(followers_timeline, indent=4))
```

```python
num_folowers = 0
for i in followers_timeline:
	num_folowers += len(followers_timeline[i])
num_folowers
```

```python
search_words = ["trophy", "skin"
                "fur",
                "dragon",
                "pelt",
                "tusk",
                "ivory",
                "scale",
                "taxidermy",
                "rug",
                "hide",
                "bone",
                "meat",
                "delicacy",
                "medicine",
                "live",
                "pangolin",
                "leopard",
                "rhino",
                "sungazer",
                "lizard",
                "crocodile",
                "alligator",
                "parrot",
                "snake",
                "python",
                "yellow material",
                "white plastic",
                "jelly",
                "aloo",
                "kola",
                "australian teddy bear",
                "stripped t-shirt",
                "four wheeler",
                "antique",
                "mammoth",
                "carved",
                "sale",
                "selling",
                "price",
                "xiangya",
                "african material",
                "ANTEBELLUMELEG1"]
tweets_with_search_words = []
# for user in followers_timeline:
    # print(user)
for follower in followers_timeline:
    # print(follower)
    try:
        tweet_timeline = followers_timeline[follower]["data"]
        for tweet in tweet_timeline:
            try:
                # print(tweet)
                tweet_text = tweet['text']
                tweet_id = tweet['id']
                # print(tweet_text)
                for search_word in search_words:
                    if search_word in tweet_text.lower():
                        # print(tweet_text)
                        bad_tweet = {}
                        bad_tweet['user'] = follower
                        bad_tweet['tweet'] = tweet_text
                        bad_tweet['id'] = tweet_id
                        tweets_with_search_words.append(bad_tweet)
            except:
                pass
    except:
        pass
followers_analysis_json['tweets_with_search_words'] = tweets_with_search_words
# print("Tweets with search words: ", tweets_with_search_words",)
```

```python
# Write user_analysis_json to json file
with open('../../data/random_users_timeline_analysis.json', 'w') as outfile:
	json.dump(followers_analysis_json, outfile, indent=4)
```
